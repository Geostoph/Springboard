```python
import pandas as pd
```

```python
import numpy as np
```

```python
import json
from pandas.io.json import json_normalize
```

```python
# Load the json file as text.
df = json.load((open('world_bank_projects.json')))

# Normalize the data in the major project theme column for easier analysis.
data = json_normalize(df, 'mjtheme_namecode')

# Look at a some rows.
data.head(15)
```

```python
top10themes = data.groupby('name').count().sort_values('code',ascending = False)
top10themes.head(10)
```

```python
data2 = data.sort_values('code')
data2
```

```python
data2[data2['name']==""] = np.NaN
```

```python
data2
```

```python
# Replace the NAs by forward filling.
data3 = data2.fillna(method='ffill')
```

```python
data3
```

```python
blanksfilledtop10=data3.groupby('name').count().sort_values('code',ascending = False)
blanksfilledtop10.head(10)
```
