## Questions 2 & 3
Finding the Top 10 Project Themes and replacing any missing values in the Name column.

To answer this question we will load the json file as text instead of the previous method.

```python
# Load the json file as text.
df = json.load((open('world_bank_projects.json')))

# Normalize the data in the major project theme column for easier analysis.
easy = json_normalize(df, 'mjtheme_namecode')

# Look at a some rows.
easy.head()
```

To answer this question we need a complete dataset.
So first we have to fill any blanks with NA.
Then we need to replace the NAs with values from the row above.
Finally, we can count and sort to answer find the 10 ten themes.

```python
# Fill missing cells with NaNs.
easy[easy['name'] == ''] = np.nan
```

```python
# Replace the NAs by forward filling.
easy = easy.fillna(method='ffill')
```

```python
# Chain methods to 
final = easy.groupby('name').count().sort_values('code',ascending =False)
final.head(10)
```

The Top 10 project themes in descending order are shown in the table above.
